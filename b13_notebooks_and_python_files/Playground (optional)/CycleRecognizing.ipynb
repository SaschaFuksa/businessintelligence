```python
import stumpy
matches = stumpy.match(dataframe_pattern['magnetometer_z_median'], dataframe['magnetometer_z_median'])
matches
```

```python
matches = matches[matches[:, 0] < 10]
len(matches)
```

```python
sorted_matches = sorted(matches, key=lambda x: x[1])
```

```python
match_df = pandas.DataFrame(columns=['near', 'index'])
index = 0
while index < len(sorted_matches):
    near = sorted_matches[index][0]
    global_index = sorted_matches[index][1]
    '''match_distance = None
    if index+1 < len(sorted_matches):
        match_distance = sorted_matches[index+1][1] - sorted_matches[index][1]
    else:
        match_distance = dataframe.index[-1] - sorted_matches[index][1]'''
    new_row = {'near':near, 'index':global_index}
    match_df = match_df.append(new_row, ignore_index=True)
    index += 1
match_df
```
