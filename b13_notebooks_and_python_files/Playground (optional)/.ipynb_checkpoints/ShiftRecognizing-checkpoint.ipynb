```python
import pandas
path = 'result/data_inside_20211017_160006.pkl.csv'
```

```python
dataframe = pandas.read_csv(path) 
dataframe = dataframe.rename(columns={'Unnamed: 0':'timer'})
dataframe = dataframe.set_index(dataframe.timer)
dataframe = dataframe.drop('timer', axis=1)
dataframe
```

```python
date = dataframe.index[0]
dates = date.split()
date = dates[0]
date
```

```python
dataframe['shift'] = 3
dataframe['shift'].mask((dataframe.index >= date + ' 06:00:00.000') & (dataframe.index < date + ' 14:00:00.000'), 1, inplace=True)
dataframe['shift'].mask((dataframe.index >= date + ' 14:00:00.000') & (dataframe.index < date + ' 22:00:00.000'), 2, inplace=True)
dataframe
```
