```python
import pandas
import numpy
import stumpy
from datetime import datetime
import matplotlib.pyplot as plt

path = 'WI3_BusinessIntelligence_Data_Bi3/data_outside_20211017_160006.pkl.csv'
#path = 'WI3_BusinessIntelligence_Data_Bi3/data_outside_20211028_000102.pkl.csv'
ref_path = 'WI3_BusinessIntelligence_Data_Bi3/data_outside_20211018_000100.pkl.csv'
```

```python
#dataframe = pandas.read_csv(path, usecols=['magnetometer_z_median', 'Unnamed: 0'])
dataframe = pandas.read_csv(path)
dataframe = dataframe.rename(columns={'Unnamed: 0':'timer'})
dataframe.dtypes
```

```python
dataframe['timer'] = dataframe['timer'].astype('datetime64[ns]')
dataframe.dtypes
```

```python
date = dataframe.timer.iloc[0]
date = datetime.strftime(date, '%Y-%m-%d')
date
```

```python
dataframe['shift'] = 3
dataframe
```

```python
#dataframe['timer'] = dataframe['timer'].astype('string')
#dataframe.dtypes
```

```python
dataframe['shift'].mask((dataframe.timer >= date + ' 06:00:00.000') & (dataframe.timer < date + ' 14:00:00.000'), 1, inplace=True)
dataframe['shift'].mask((dataframe.timer >= date + ' 14:00:00.000') & (dataframe.timer < date + ' 22:00:00.000'), 2, inplace=True)
dataframe
```

```python
dataframe_ref = pandas.read_csv(ref_path, usecols=['magnetometer_z_median', 'Unnamed: 0']) 
dataframe_ref = dataframe_ref.rename(columns={'Unnamed: 0':'timer'})
#dataframe_ref = dataframe_ref.loc[(dataframe_ref.index >= 253) & (dataframe_ref.index <= 625)]
dataframe_ref = dataframe_ref.loc[(dataframe_ref.index >= 106) & (dataframe_ref.index <= 478)]
dataframe_ref
```

```python
plt.style.use('https://raw.githubusercontent.com/TDAmeritrade/stumpy/main/docs/stumpy.mplstyle')
plt.plot(dataframe_ref['magnetometer_z_median'].values)
plt.show()
```

```python
matches = stumpy.match(dataframe_ref['magnetometer_z_median'], dataframe['magnetometer_z_median'])
len(matches)
```

```python
#Entferne alle Werte mit Ähnlichkeit < 10
#matches = matches[matches[:, 0] < 10]
#len(matches)
```

```python
matches = sorted(matches, key=lambda x: x[1])
match_df = pandas.DataFrame(matches, columns=['similarity', 'index'])
match_df
```

```python
match_df['distance'] = match_df['index'].shift(-1).sub(match_df['index'])
match_df
```

```python
last_index = int(match_df['index'].iloc[-1])
last_index
```

```python
minindex = dataframe.magnetometer_z_median[last_index+200:last_index+400].idxmin()
minindex
```

```python
match_df['distance'].iloc[-1] = minindex-last_index
match_df['distance'] = match_df['distance'].astype('int')
match_df
```

```python
match_df.nlargest(10, ['distance'])
```

```python
match_df.nsmallest(10, ['distance'])
```

```python
unexpected_distances = match_df[match_df['distance'] > 400]
unexpected_distances
```

```python
for row in unexpected_distances.itertuples():
    start_index = row.index
    end_index = dataframe.magnetometer_z_median[start_index+200:start_index+400].idxmin()
    match_df['distance'].iloc[row.Index] = end_index-start_index
match_df.nlargest(10, ['distance'])
```

```python
# Add cycle to df
merged_dataframe = dataframe.copy()
merged_dataframe['cycle'] = numpy.nan
#for row in match_df_cleaned.itertuples():
cycle_number = 1
for row in match_df.itertuples():
    start_index = row.index
    end_index = start_index + row.distance
    merged_dataframe.loc[start_index:end_index, 'cycle'] = cycle_number
    cycle_number += 1
merged_dataframe
```

```python
merged_dataframe = merged_dataframe.dropna()
merged_dataframe['cycle'] = merged_dataframe['cycle'].astype('int')
merged_dataframe
```

```python
# Prod Link
production_path = 'WI3_BusinessIntelligence_Result_Bi3/production.csv'
production_dataframe = pandas.read_csv(production_path, usecols=['timer', 'Protokollzyklenzähler', 'Zykluszeit, Istwert'])
production_dataframe = production_dataframe.rename(columns={'Protokollzyklenzähler':'cycle_number'})
production_dataframe['cycle_number'] = production_dataframe['cycle_number'].astype('int')
production_dataframe = production_dataframe.rename(columns={'Zykluszeit, Istwert':'cycle_length'})
production_dataframe
```

```python
production_dataframe['timer'] = production_dataframe['timer'].astype('datetime64[ns]')
production_dataframe.dtypes
```

```python
start_date = dataframe.timer.iloc[0]
start_date = datetime.strftime(start_date, '%Y-%m-%d')
start_date
```

```python
#end_date = start_date + timedelta(days=1)
end_date = pandas.to_datetime(start_date) + pandas.DateOffset(days=1)
end_date = datetime.strftime(end_date, '%Y-%m-%d')
end_date
```

```python
production_dataframe = production_dataframe.loc[(production_dataframe.timer > start_date) & (production_dataframe.timer < end_date)]
production_dataframe
```

```python
match_df_time = match_df.copy()
match_df_time['start_time'] = numpy.nan
match_df_time['end_time'] = numpy.nan
match_df_time
```

```python
for row in match_df_time.itertuples():
    match_df_time.start_time[row.Index] = dataframe.timer.iloc[int(row.index)]
    end_index = int(row.index)+int(row.distance)-1
    match_df_time.end_time[row.Index] = dataframe.timer.iloc[end_index]
match_df_time
```

```python
match_df_prod_id = match_df_time.copy()
match_df_prod_id['prod_id'] = numpy.nan
match_df_prod_id
```

```python
# Dataframe auf Minute kürzen, und dann mit Groupby schauen
merged_df_timer_min = merged_dataframe.copy()
merged_df_timer_min['timer_minute'] = merged_df_timer_min.timer.dt.round('min')
merged_df_timer_min[600:1200]
```

```python
merged_df_timer_sec = merged_df_timer_min.copy()
merged_df_timer_sec['timer_second'] = merged_df_timer_sec.timer.dt.round('s')
merged_df_timer_sec
```

```python
grouped_df = merged_df_timer_sec.groupby("timer_minute")
grouped_lists = grouped_df["cycle"].apply(set)
grouped_lists = grouped_lists.reset_index()
grouped_lists = grouped_lists.set_index('timer_minute')
grouped_lists
```

```python
grouped_df_prod = production_dataframe.groupby("timer")
grouped_lists_prod = grouped_df_prod["cycle_number"].apply(set)
grouped_lists_prod = grouped_lists_prod.reset_index()
grouped_lists_prod = grouped_lists_prod.set_index('timer')
grouped_lists_prod
```

```python
merged_cycles = pandas.concat([grouped_lists, grouped_lists_prod], axis=1)
merged_cycles.head(20)
```

```python
mapping_cycles = pandas.DataFrame(columns=['cycle', 'cycle_number'])
for row in merged_cycles.itertuples():
    cycles = row.cycle
    cycle_numbers = row.cycle_number
    cycle_index = 0
    if pandas.notnull(cycles) and pandas.notnull(cycle_numbers):
        cycles = sorted(cycles)
        cycle_numbers = sorted(cycle_numbers)
        for cycle_number in cycle_numbers:
            if len(cycles) > cycle_index:
                new_row = {'cycle': cycles[cycle_index], 'cycle_number': cycle_number}
                mapping_cycles = mapping_cycles.append(new_row, ignore_index=True)
                cycle_index += 1
mapping_cycles
```

```python
merged_df_with_cnr = merged_df_timer_sec.copy()
merged_df_with_cnr['cycle_number'] = numpy.nan
for row in mapping_cycles.itertuples():
    merged_df_with_cnr.cycle_number.loc[merged_df_with_cnr['cycle'] == row.cycle] = row.cycle_number
merged_df_with_cnr['cycle_number'] = merged_df_with_cnr['cycle_number'].fillna(0).astype('int')
merged_df_with_cnr
```

```python
merged_df_with_cnr[500:600]
```

```python
prod_dataframe_dup = production_dataframe.copy()
prod_dataframe_dup['has_duplicate'] = prod_dataframe_dup.duplicated(subset=['timer'])
prod_dataframe_dup
```

```python
prod_dataframe_linked = prod_dataframe_dup.copy()
prod_dataframe_linked['timer_start'] = numpy.nan
prod_dataframe_linked['timer_end'] = numpy.nan
for row in prod_dataframe_linked.itertuples():
    if row.Index-1 in prod_dataframe_linked.index and ((row.has_duplicate) or (row.timer <= prod_dataframe_linked.timer_end[row.Index-1])):
        prod_dataframe_linked.timer_start[row.Index] = prod_dataframe_linked.timer_end[row.Index-1]
        prod_dataframe_linked.timer_end[row.Index] = prod_dataframe_linked.timer_start[row.Index] + pandas.to_timedelta(row.cycle_length, unit='s')
    else:
        prod_dataframe_linked.timer_start[row.Index] = row.timer
        prod_dataframe_linked.timer_end[row.Index] = row.timer + pandas.to_timedelta(row.cycle_length, unit='s')
prod_dataframe_linked.head(10)
```

```python
gr = prod_dataframe_linked.groupby(prod_dataframe_linked.timer, sort=True).size().sort_values(ascending=False)
gr
```
