```python
import pandas
import numpy
import datetime

path = 'result/data_outside_20211017_160006.pkl.csv'
```

```python
dataframe = pandas.read_csv(path) 
dataframe = dataframe.rename(columns={'Unnamed: 0':'timer'})
#dataframe = dataframe.set_index(dataframe.timer)
#dataframe = dataframe.drop('timer', axis=1)
dataframe
```

```python
date = dataframe.timer[0]
dates = date.split()
date = dates[0]
#dataframe = dataframe.loc[(dataframe.timer >= date+' 18:40') & (dataframe.timer < date+' 18:41')]
#dataframe = dataframe.reset_index()
#dataframe
```

```python
dataframe['cycle'] = None
dataframe
```

```python
start_time = dataframe.timer[0]
end_time = dataframe.timer.iloc[-1]
i = 0
df = pandas.DataFrame(columns=['timer', 'index', 'value'])
while i+400 < dataframe.index[-1]:
    part = dataframe[i+200:i+400]
    minindex = part['magnetometer_z_median'].idxmin()
    #print(minindex)
    minvalue = part['magnetometer_z_median'].min()
    #print(minvalue)
    i = minindex
    start_time = dataframe.timer[minindex]
    #print(start_time)
    new_row = {'timer':start_time, 'index':minindex, 'value':minvalue}
    df = df.append(new_row, ignore_index=True)
df
```

```python
df2 = df.drop(df[df.value > -60000000.0].index)
df2
```
