```python
#https://ourcodingclub.github.io/tutorials/pandas-time-series/
#https://towardsdatascience.com/how-to-painlessly-analyze-your-time-series-f52dab7ea80d
#https://matrixprofile.org/
m = 640
mp = stumpy.stump(dataframe['magnetometer_x_mean'], m)
mp
```

```python
print(len(mp))
```

```python
motif_idx = numpy.argsort(mp[:, 0])[0]
motif_idx
```

```python
nearest_neighbor_idx = mp[motif_idx, 1]
nearest_neighbor_idx
```

```python
fig, axs = plt.subplots(2, sharex=True, gridspec_kw={'hspace': 0})
plt.suptitle('Motif (Pattern) Discovery', fontsize='30')

axs[0].plot(dataframe['magnetometer_x_mean'].values)
axs[0].set_ylabel('magnetometer_x_mean', fontsize='20')
rect = Rectangle((motif_idx, 0), m, 40, facecolor='lightgrey')
axs[0].add_patch(rect)
rect = Rectangle((nearest_neighbor_idx, 0), m, 40, facecolor='lightgrey')
axs[0].add_patch(rect)

axs[1].set_ylabel('Matrix Profile', fontsize='20')
axs[1].axvline(x=motif_idx, linestyle="dashed")
axs[1].axvline(x=nearest_neighbor_idx, linestyle="dashed")
axs[1].plot(mp[:, 1])
plt.show()
```
